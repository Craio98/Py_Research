```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import integrate

#DATE INITIALE 
#Pozitie .CSV / .UHS . Locatie folder ori url 
fileName="https://github.com/Craio98/Py_Research/raw/main/graph.uhs"
#Densitatea materialului acceptor in g/cm^3
ro="1.50"

#Executare algoritm
myCSV=pd.read_csv(fileName,skiprows=4, names=["Number", "Time", "Gauges"])
df=pd.DataFrame(myCSV)
#Delete lower values (lower than ~1atm)  
selector=df[(df['Gauges'] <= 15000)].index
df.drop(selector,inplace=True)


myGraph=plt.plot(df['Time'],df['Gauges'] )
plt.xlabel("Timp [ms]")
plt.ylabel("Presiune [kPa]")
plt.legend(["Senzor 1"], loc='best',title="Legenda")
print(df)
```

```python
poly = np.polyfit(df['Time'],df['Gauges'], deg=4)
print(poly)
```

```python
myPol=lambda x: poly[1]*x**4 + poly[2]*x**3+poly[3]*x**2+poly[4]*x
myIntegral=integrate.quad(myPol, 0, 4.5)
```

```python
Pmax=df['Gauges'].max()
```

```python
E=Pmax*myIntegral[1]/(2250*1.50)
print(E)
```
